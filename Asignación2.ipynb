```python
# LECTURA DE DATOS

import pandas as pd
import numpy as np
from datetime import datetime

df = pd.read_csv('Smartphone_Evolution.csv')
print(df.info())
df
```

```python
# Primero, obtenga un resumen del conjunto de datos con estadísticas básicas como recuento, media, estándar, mínimo y máximo.
print(df.describe())
```

```python
# La cantidad de teléfonos inteligentes lanzados cada año para comprender la tendencia a lo largo del tiempo.
import matplotlib.pyplot as plt

df['Year'] = pd.to_datetime(df['Release_Date']).dt.year
yearly_counts = df['Year'].value_counts().sort_index()

plt.figure(figsize=(12, 6))
plt.plot(yearly_counts.index, yearly_counts.values, marker='o')
plt.title('Número de smartphones lanzados por año')
plt.xlabel('Año')
plt.ylabel('Número de Smartphones')
plt.grid(True)
plt.show()
```

```python
# Principales marcas de teléfonos inteligentes por la cantidad de modelos que han lanzado.

top_brands = df['Brand'].value_counts().head(10)

plt.figure(figsize=(12, 6))
top_brands.plot(kind='bar')
plt.title('Top 10 Marcas de Smartphone por cantidad de modelos lanzados')
plt.xlabel('Marca')
plt.ylabel('Numero de Modelos')
plt.xticks(rotation=45)
plt.show()
```

```python
# Distribución de tamaños de pantalla para comprender los tamaños de pantalla comunes utilizados en los teléfonos inteligentes.

plt.figure(figsize=(12, 6))
df['Display_Size'].dropna().plot(kind='hist', bins=30, edgecolor='black')
plt.title('Distribución de tamaños de pantalla')
plt.xlabel('Tamaño de pantalla (pulgadas)')
plt.ylabel('Frecuencia')
plt.grid(True)
plt.show()
```

```python
# Analizar la capacidad de la batería de los teléfonos inteligentes a lo largo de los años.

# Convertir la 'Capacidad de la batería' a numérica, forzando errores a NaN y eliminando filas con NaN
df['Battery'] = pd.to_numeric(df['Battery'], errors='coerce')
df_cleaned = df.dropna(subset=['Battery', 'Year'])

# Capacidad de bateria por año
average_battery_capacity = df_cleaned.groupby('Year')['Battery'].mean()

# Plot the trend
plt.figure(figsize=(12, 6))
plt.plot(average_battery_capacity.index, average_battery_capacity.values, marker='o', linestyle='-')
plt.title('Capacidad promedio en el tiempo')
plt.xlabel('Año')
plt.ylabel('Capacidad promedio (mAh)')
plt.grid(True)
plt.show()
```
